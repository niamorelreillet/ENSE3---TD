## Defining a function (two ways)

```python
def myFunction(a,b):
    return (a+b,a-b)
(plus,minus)=myFunction(1.0,3.0)
print(plus,minus)

myLambda=lambda a,b: (a+b,a-b)
(plus,minus)=myLambda(1.0,3.0)
print(plus,minus)
```

## X-Y plot (with subplots)

```python
from math import pi
x=np.linspace(0, 2.0*pi, num=30)
y=np.sin(x)
z=np.exp(x)

plt.rcParams["figure.figsize"] = (7,7)
fig,ax=plt.subplots(2,2,constrained_layout=True)

cax = ax[0,0]
csf = cax.plot(x,y)

cax = ax[1,0]
csf = cax.plot(x,y, color='green', linestyle='dashed', marker='o',
     markerfacecolor='blue', markersize=8)

cax = ax[0,1]
csf = cax.plot(x,y, marker="x")
cax.set_xlabel("x",fontsize=14)
cax.set_ylabel("sin(x)",fontsize=14)
cax.grid(True)

cax = ax[1,1]
csf = cax.plot(x,z, marker="x")
cax.set_yscale('log')
cax.set_xlabel("x",fontsize=14)
cax.set_ylabel("exp(x)",fontsize=14)
```

## Finding the root (in a given interval a univariate function (Brent method)

```python
from scipy.optimize import brentq

f=lambda x: (x-2)*x
root1=brentq(f, -1, 1)
print(root1)
root2=brentq(f, 1, 3)
print(root2)
```
